```python
data['FLOODS'].replace(['YES','NO'],[1,0],inplace=True)
```

```python
data.head()
```

```python
x=data.iloc[:,2:14] #features/columns
x.head()
```

```python
y=data.iloc[:,-1] #target or labels
y
```

```python
c=data[['JUN','JUL','AUG','SEP']]
c.hist(bins=40,figsize=(15,10))
plt.show()
```

```python
data[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].mean().plot.\
            bar(figsize=(14,6))
plt.xlabel('Month')
plt.ylabel('Monthly Rainfall')
plt.show()
```

```python
from sklearn import preprocessing
minmax=preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier #packages for decision tree
dtc_clf=DecisionTreeClassifier(criterion = "entropy") #model creation
dtc_clf.fit(x_train,y_train) #model training
```

```python
y_pred=dtc_clf.predict(x_test)
y_pred
```

```python
pd.DataFrame({'actual':y_test,'prediction':y_pred})
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=200)
clf.fit(x_train,y_train)
```

```python
y_pred2=clf.predict(x_test)
y_pred2
```

```python
from sklearn import metrics
print("Accuracy of Random Forest:",metrics.accuracy_score(y_test, y_pred2)*100)
```
